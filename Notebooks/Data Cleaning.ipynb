```python
# Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_rows = 20
np.set_printoptions(precision = 4, suppress = True)
```

```python
# Children under 4 dying due to HIV/AIDS

file = "AIDS-HIV.csv"

HIV_df = pd.read_csv(file)

HIV_df.drop([0,1,2], axis = 0, inplace = True)

# TODO Try put into loop

HIV_df['2017'] = HIV_df['2017'].astype(float) + HIV_df['2017.1'].astype(float) + HIV_df['2017.2'].astype(float)
HIV_df['2016'] = HIV_df['2016'].astype(float) + HIV_df['2016.1'].astype(float) + HIV_df['2016.2'].astype(float)
HIV_df['2015'] = HIV_df['2015'].astype(float) + HIV_df['2015.1'].astype(float) + HIV_df['2015.2'].astype(float)
HIV_df['2014'] = HIV_df['2014'].astype(float) + HIV_df['2014.1'].astype(float) + HIV_df['2014.2'].astype(float)
HIV_df['2013'] = HIV_df['2013'].astype(float) + HIV_df['2013.1'].astype(float) + HIV_df['2013.2'].astype(float)
HIV_df['2012'] = HIV_df['2012'].astype(float) + HIV_df['2012.1'].astype(float) + HIV_df['2012.2'].astype(float)
HIV_df['2011'] = HIV_df['2011'].astype(float) + HIV_df['2011.1'].astype(float) + HIV_df['2011.2'].astype(float)
HIV_df['2010'] = HIV_df['2010'].astype(float) + HIV_df['2010.1'].astype(float) + HIV_df['2010.2'].astype(float)
HIV_df['2009'] = HIV_df['2009'].astype(float) + HIV_df['2009.1'].astype(float) + HIV_df['2009.2'].astype(float)
HIV_df['2008'] = HIV_df['2008'].astype(float) + HIV_df['2008.1'].astype(float) + HIV_df['2008.2'].astype(float)
HIV_df['2007'] = HIV_df['2007'].astype(float) + HIV_df['2007.1'].astype(float) + HIV_df['2007.2'].astype(float)

HIV_df.drop(['2006', '2006.1', '2006.2', '2005', '2005.1', '2005.2',
             '2004', '2004.1', '2004.2', '2003', '2003.1', '2003.2', 
             '2002', '2002.1', '2002.2', '2001', '2001.1', '2001.2',
             '2000', '2000.1', '2000.2', '2017.1', '2017.2', '2016.1',
             '2016.2', '2015.1', '2015.2', '2014.1', '2014.2', '2013.1',
             '2013.2','2012.1', '2012.2', '2012.1', '2012.2', '2011.1',
             '2011.2', '2010.1', '2010.2', '2009.1', '2009.2', 
             '2008.1', '2008.2', '2007.1', '2007.2'], axis = 1, inplace = True)



HIV_df.rename(columns = {'Unnamed: 0' : 'Country Name'}, inplace = True)
HIV_df = HIV_df[['Country Name', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']]

HIV_df.to_csv('finalhiv.csv')

HIV_df
```

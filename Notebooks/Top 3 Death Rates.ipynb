# Top 3 Death Rates

```python
# Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
pd.options.display.max_rows = 20
np.set_printoptions(precision = 4, suppress = True)
```

```python
gdp_df = pd.read_csv("finalgdp.csv")
gdp_df.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
HIV_df = pd.read_csv("finalhiv.csv")
HIV_df.drop("Unnamed: 0", axis = 1, inplace = True)
```

## Nigeria, South Africa and Mozambique

```python
# Nigeria
Nigeria_HIV = HIV_df[HIV_df['Country Name'] == 'Nigeria']
Nigeria_gdp = gdp_df[gdp_df['Country Name'] == 'Nigeria']
pivot_HIV = Nigeria_HIV.pivot_table(columns = 'Country Name')
pivot_gdp = Nigeria_gdp.pivot_table(columns = 'Country Name')
pivot_HIV.rename(columns = {'Nigeria' : 'Deaths'}, inplace = True)
pivot_gdp.rename(columns = {'Nigeria' : 'GDP'}, inplace = True)
Nigeria_df = pivot_gdp.join(pivot_HIV, how = 'outer')

# South Africa
Safrica_HIV = HIV_df[HIV_df['Country Name'] == 'South Africa']
Safrica_gdp = gdp_df[gdp_df['Country Name'] == 'South Africa']
pivot_HIV = Safrica_HIV.pivot_table(columns = 'Country Name')
pivot_gdp = Safrica_gdp.pivot_table(columns = 'Country Name')
pivot_HIV.rename(columns = {'South Africa' : 'Deaths'}, inplace = True)
pivot_gdp.rename(columns = {'South Africa' : 'GDP'}, inplace = True)
Safrica_df = pivot_gdp.join(pivot_HIV, how = 'outer')

# Mozambique
Moz_HIV = HIV_df[HIV_df['Country Name'] == 'Mozambique']
Moz_gdp = gdp_df[gdp_df['Country Name'] == 'Mozambique']
pivot_HIV = Moz_HIV.pivot_table(columns = 'Country Name')
pivot_gdp = Moz_gdp.pivot_table(columns = 'Country Name')
pivot_HIV.rename(columns = {'Mozambique' : 'Deaths'}, inplace = True)
pivot_gdp.rename(columns = {'Mozambique' : 'GDP'}, inplace = True)
Moz_df = pivot_gdp.join(pivot_HIV, how = 'outer')
```

```python
fig = figure(num=None, figsize=(8, 7), dpi = 80, facecolor = 'w', edgecolor = 'k')
fig.add_subplot(3, 1, 1)
fig.set_facecolor('#dfdfdf')

# Nigeria
Nigeria_df['Deaths'].plot(kind = 'bar', color = '#3399ff')
plt.ylabel('Deaths')
Nigeria_df['GDP'].plot(secondary_y = True, color = 'g', linewidth = '3', linestyle = '--')
plt.ylabel('GDP')
plt.title('Nigeria - GDP vs Death Count over time')
plt.legend(['GDP'])

# South Africa
fig.add_subplot(3, 1, 2)
Safrica_df['Deaths'].plot(kind = 'bar', color = '#3399ff')
plt.ylabel('Deaths')
Safrica_df['GDP'].plot(secondary_y = True, color = 'g', linewidth = '3', linestyle = '--')
plt.ylabel('GDP')
plt.title('South Africa - GDP vs Death Count over time')
plt.legend(['GDP'])

# Mozambique
fig.add_subplot(3, 1, 3)
Moz_df['Deaths'].plot(kind = 'bar', color = '#3399ff')
plt.ylabel('Deaths')
Moz_df['GDP'].plot(secondary_y = True, color = 'g', linewidth = '3', linestyle = '--')
plt.ylabel('GDP')
plt.title('Mozambique - GDP vs Death Count over time')
plt.legend(['GDP'])



#fig.savefig("top_3", bbox_inches = 'tight')
```
